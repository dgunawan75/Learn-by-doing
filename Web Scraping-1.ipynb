```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
from pandas import Series,DataFrame
```

```python
url ='https://www.ucop.edu/operating-budget/budgets-and-reports/2024-25-legislative-session.html'
```

```python
# Request content from web page
result = requests.get(url)
c = result.content

# Set as Beautiful Soup Object
soup = BeautifulSoup(c)
```

```python
# Go to the section of interest
summary = soup.find("div",{'class':'list-land','id':'content'})

# Check if the summary div was found
if summary is None:
    print("Error: Could not find the 'summary' div with class 'list-land' and id 'content'.")
    tables = [] # Initialize tables as an empty list to prevent further errors
else:
    # Find the tables in the HTML
    tables = summary.find_all('table')
    print (tables)
```

```python
# Set up empty data list
data = []

# Set rows as first indexed object in tables with a row
rows = tables[0].findAll('tr')

# now grab every HTML cell in every row
for tr in rows:
    cols = tr.findAll('td')
    # Check to see if text is in the row
    for td in cols:
        text = td.find(text=True)
        print (text),
        data.append(text)
```

```python
data
```

```python
# Set up empty lists
reports = []
date = []

# Se tindex counter
index = 0

# Go find the pdf cells
for item in data:
    if 'pdf' in item:
        # Add the date and reports
        date.append(data[index-1])

        # Get rid of \xa0
        reports.append(item.replace(u'\xa0', u' '))

    index += 1
```

```python
# Set up Dates and Reports as Series
date = Series(date)
reports = Series(reports)
```

```python
# Concatenate into a DataFrame
legislative_df = pd.concat([date,reports],axis=1)
```

```python
# Set up the columns
legislative_df.columns = ['Date','Reports']
```
